### 1085. 직사각형에서 탈출

```python
x, y, w, h = map(int, input().split())

print(min((w-x), (h-y), x, y))
```

### 4153. 직각삼각형

```python
while True:
    nums = list(map(int, input().split()))
    nums.sort()
    a, b, c = nums[0], nums[1], nums[2]

    if a == 0 and b == 0 and c == 0:
        break
    elif (a**2 + b**2) == c**2:
        print('right')
    else:
        print('wrong')
```

### 10250. ACM 호텔

```python
T = int(input())

for _ in range(T):
    h, w, n = map(int, input().split())
    cnt = 0

    for i in range(w):
        for j in range(h):
            cnt += 1

            if cnt == n:
                print('%d%02d'%(j+1, i+1))
                break
```
